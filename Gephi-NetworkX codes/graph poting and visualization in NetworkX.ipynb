```
import pandas as pd
df = pd.read_csv('OTCNet.txt', sep=',')  #   , skiprows=1
```

```
print (df)
```

```
G2 = nx.DiGraph()
```

```
for idx, row in df.iterrows():
    G2.add_node(row['From'], to=row['To'], trust=row['Trust'])
```

```
trust1 = [ d['trust'] for n, d in G2.nodes(data=True)]
```

```
nx.draw_networkx(G2, node_color=trust1)
```

```
G11=nx.read_weighted_edgelist("OTCNet.csv",delimiter=',', nodetype=int)
```

```
G11
```

```
pos = nx.spring_layout(G11)
```

```
# nodes
nx.draw_networkx_nodes(G11, pos, node_size=5)
```

```
elarge = [(u, v) for (u, v, d) in G11.edges(data=True) if d['weight'] > 0.5]
esmall = [(u, v) for (u, v, d) in G11.edges(data=True) if d['weight'] <= 0.5]
```

```
# edges
nx.draw_networkx_edges(G11, pos, edgelist=elarge,
                       width=6)
nx.draw_networkx_edges(G11, pos, edgelist=esmall,
                       width=6, alpha=0.5, edge_color='b', style='dashed')
```

```
# labels
nx.draw_networkx_labels(G11, pos, font_size=20, font_family='sans-serif')
```

```
plt.axis('off')
```
